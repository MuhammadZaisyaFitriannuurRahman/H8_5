```python
path_sales = 'dataset/nyc-rolling-sales.csv'
```

```python
df_sales = pd.read_csv(path_sales)
```

```python
# Drop kolom yang tidak diperlukan
column_drop = ['Unnamed: 0']
df_sales.drop(column_drop, axis=1, inplace=True)
```

```python
df_sales
```

```python
# Konversi 'SALE PRICE' menjadi tipe data numerik
df_sales['SALE PRICE'] = pd.to_numeric(df_sales['SALE PRICE'], errors='coerce')

# Memfilter data dengan jumlah dolar yang tidak masuk akal
df_sales = df_sales[df_sales['SALE PRICE'] > 0]
```

```python
# Memilih kolom yang relevan
df_sales = df_sales[['BOROUGH', 'TOTAL UNITS']]
```

```python
# Menghitung Central Tendency
mean_units_sold = df_sales['TOTAL UNITS'].mean()
median_units_sold = df_sales['TOTAL UNITS'].median()
mode_units_sold = df_sales['TOTAL UNITS'].mode()[0]
range_units_sold = df_sales['TOTAL UNITS'].max() - df_sales['TOTAL UNITS'].min()
variance_units_sold = df_sales['TOTAL UNITS'].var()
std_dev_units_sold = df_sales['TOTAL UNITS'].std()
```

Rerata (Mean): Rerata unit terjual adalah jumlah rata-rata dari total unit yang terjual.

Median: Median unit terjual adalah nilai tengah dari total unit yang terjual, diurutkan dari terkecil hingga terbesar.

Modus: Modus unit terjual adalah nilai atau kategori yang paling sering muncul dalam data.

Jangkauan (Range): Jangkauan unit terjual adalah selisih antara nilai terbesar dan terkecil dari total unit yang terjual.

Varians: Varians unit terjual adalah pengukuran sebaran data yang mengukur seberapa jauh data tersebar dari rerata.

Deviasi Standar: Deviasi standar unit terjual adalah akar kuadrat dari varians dan mengukur sebaran data dalam distribusi.

```python
# Step 4: Lakukan one-way ANOVA test
from scipy.stats import f_oneway
```

```python
boroughs = df_sales['BOROUGH'].unique()
data_by_borough = [df_sales[df_sales['BOROUGH'] == b]['TOTAL UNITS'] for b in boroughs]
f_statistic, p_value = f_oneway(*data_by_borough)
```

```python
alpha = 0.05
if p_value < alpha:
    conclusion = "Menolak hipotesis nol: Terdapat perbedaan signifikan dalam unit yang terjual antara boroughs."
else:
    conclusion = "Tidak menolak hipotesis nol: Tidak terdapat perbedaan signifikan dalam unit yang terjual antara boroughs."
```

```python
# Print hasil
print(f"Mean units sold: {mean_units_sold}")
print(f"Median units sold: {median_units_sold}")
print(f"Mode units sold: {mode_units_sold}")
print(f"Range units sold: {range_units_sold}")
print(f"Variance units sold: {variance_units_sold}")
print(f"Standard Deviation units sold: {std_dev_units_sold}")
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
print(conclusion)
```
